```python
x = np.arange(0,10*np.pi,0.1)
y = np.sin(x)
y1 = np.cos(x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,linewidth=3,linestyle="-",color='r')
plt.plot(x,y1,linewidth=3,linestyle='-',color='green')
plt.title("sin(x)")
plt.xlabel("radian")
plt.ylabel("sin(x) and cos(x)")
plt.legend(['sin(x)','cos(x)'],loc=3,fontsize='x-large')
plt.grid()
plt.savefig("sin(x)")
plt.show()
```
